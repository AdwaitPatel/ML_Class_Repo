```python
# Linear Regression

# Equation => y = mx + c
''' 
y => dependent variable(output) 
m => slope
x => independent variable(input)
c => intercept
'''

# Y = B0 + B1X
'''
B0 => Intercept
B1 => Slope
'''
```

$$
\beta_0 = \frac{ (\sum y)(\sum x^2) - (\sum y)(\sum xy) }
                { n(\sum x^2) - (\sum x)^2 }
$$

$$
\beta_1 = \frac{ n(\sum xy) - (\sum x)(\sum y) }
                { n(\sum x^2) - (\sum x)^2 }
$$

Suppose u want to predict the weight gain based upon calories consumed only based on the below given data :

| Calories Consumed (X) | Weight Gain (Y) |
|------------------------|-----------------|
| 500                    | 1               |
| 1000                   | 2               |
| 1500                   | 2.3             |
| 1750                   | 3.4             |
| 2100                   | 3.4             |


Suppose we want to predict the weight gain when u consume  2500 calories and 3000 calories

```python
# Simple Linear Regression python code.

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

#1. Prepare the data
# Independent variable (features) - must be a 2D array for scikit-learn
X = np.array([500, 1000, 1500, 1750, 2100]).reshape(-1, 1)
# Dependent variable (target)
y = np.array([1, 2, 2.5, 3.2, 3.9])

#2. Create and train the model
model = LinearRegression()
model.fit(X, y)

#3. Get the model's coefficients
intercept = model.intercept_
coefficient = model.coef_[0]

print(f"Intercept (b0): {intercept}")
print(f"Coefficient (b1): {coefficient}")

#4. Make predictions
y_pred = model.predict(X)

#5. Visualize the results (optional)
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (y)")
plt.title("Simple Linear Regression")
plt.legend()
plt.grid(True)
plt.show()

 
#6 Prediction of the new data
new_data =  np.array([2500, 3000]).reshape(-1, 1)
predicted_value = model.predict(new_data)
print(f"Prediction for new data {new_data}: {predicted_value}")


print("\n\nNew Plot with new predicted data")

X = np.array([500, 1000, 1500, 1750, 2100, 2500, 3000]).reshape(-1, 1)
y = np.array([1, 2, 2.5, 3.2, 3.9, 4.49678841, 5.37147355])

y_pred = model.predict(X)

plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (y)")
plt.title("Simple Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```
