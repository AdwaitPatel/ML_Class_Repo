```python
import pandas as pd
import matplotlib.pyplot as plt

iris = pd.read_csv('https://raw.githubusercontent.com/venky14/Machine-Learning-with-Iris-Dataset/refs/heads/master/Iris.csv')

# print(iris)

mean = iris["PetalWidthCm"].mean()
median = iris["PetalWidthCm"].median()
mode = iris["PetalWidthCm"].mode().iloc[0]

print(f'The mean of petal width is => {mean}')
print(f'The median of petal width is => {median}')
print(f'The mode of petal width is => {mode}')

iris["PetalWidthCm"] = mean

print(iris)

```

```python
# Imputation

cols = iris.select_dtypes(include=['float64', 'int64']).columns
iris_z = iris.copy()
iris_z[cols] = 0

mean = iris[cols].mean()
mean_imputed = iris_z.copy()
mean_imputed[cols] = mean_imputed[cols].replace(0, mean)

difference = mean_imputed[cols] - iris[cols]

plt.figure(figsize=(12, 6))
for i, col in enumerate(cols, 1):
    plt.subplot(2, 3, i)
    plt.hist(difference[col], bins=15, color='pink', edgecolor='red')
    plt.title(f'Difference: {col}')
    plt.xlabel('Value Difference')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
species_counts = iris['Species'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Distribution of Iris Species')
plt.axis('equal')
plt.show()
```
